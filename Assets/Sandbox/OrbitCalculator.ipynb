```python
import matplotlib.pyplot as plt
import math
import numpy

fig = plt.figure()
ax = fig.add_subplot()
ax.set_aspect('equal', adjustable='box')

stepsize = 0.01
radius = 8
amplitude = 1
cycles = 10
phaseShift = 2

for i in numpy.arange(0,2*math.pi,stepsize):
    x = (radius + amplitude * math.sin(cycles*i+phaseShift)) * math.cos(i)
    y = (radius + amplitude * math.sin(cycles*i+phaseShift)) * math.sin(i)
    plt.scatter(x, y, s=5)

plt.show()
```
